# 📱 Flutter Engineer — Mobile App (Upload + Real-Time)
### Ishara Arabic Sign Language Translator | Graduation Project

**مسؤوليتك:**  
أنت بتبني الـ Flutter mobile app اللي المستخدم بيتعامل معاها.  
الـ app عندها mode اتنين: **Upload** (ترفع فيديو) و **Real-Time** (تصوير مباشر).

---
**الـ Flow:**
```
User opens app
    ↓
Choose mode: [Upload Video] or [Real-Time Camera]
    ↓
UPLOAD MODE:               REAL-TIME MODE:
Pick video file        |   Open camera
    ↓                  |       ↓
POST /translate/upload |   Stream frames via WebSocket
    ↓                  |       ↓
Display gloss + text   |   Display live gloss + text
```

> **dependency:** محتاج الـ **API URL + response schemas** من **Backend Engineer**
